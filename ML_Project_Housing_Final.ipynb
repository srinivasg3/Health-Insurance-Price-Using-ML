# Dealing with Outliers

```python
# Here I am using IQR to remove outliers
# IQR for BMI column
Q1 = price_new['BMI'].quantile(0.25)
Q3 = price_new['BMI'].quantile(0.75)
print(Q3, Q1)
IQR = Q3 - Q1
print(IQR)
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(upper_bound)
print(lower_bound)
outlier =[]
for x in price_new['BMI']:
    if ((x> upper_bound) or (x<lower_bound)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
```

```python
# capping method for treatment of outlier for bmi
# if bmi values are greater than upper_bound then they will be set to upper_bound value
# if bmi values are lesser than lower bound then they will be adjusted to lower_bound value
price_new['BMI'] = np.where(price_new['BMI']>upper_bound,upper_bound,np.where(price_new['BMI']<lower_bound,lower_bound,price_new['BMI']))
```

```python
# IQR for health_insurance_price column
Q1 = price_new['health_insurance_price'].quantile(0.25)
Q3 = price_new['health_insurance_price'].quantile(0.75)
print(Q3, Q1)
IQR = Q3 - Q1
print(IQR)
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(upper_bound)
print(lower_bound)
outlier =[]
for x in price_new['health_insurance_price']:
    if ((x> upper_bound) or (x<lower_bound)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
```

```python
# capping method for outliers in health insurance price
price_new['health_insurance_price'] = np.where(price_new['health_insurance_price']>upper_bound,upper_bound,
np.where(price_new['health_insurance_price']<lower_bound,lower_bound,price_new['health_insurance_price']))
```

Building The Model


```python
X_lin=price_new.drop(['health_insurance_price'],axis=1)
X_lin
```

```python
Y_lin=price_new['health_insurance_price']
Y_lin
```

```python
X_lin.shape
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_lin,Y_lin,test_size=0.2,random_state=0)
```

```python
#performing transformation of data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_train
```

```python
x_test=sc.transform(x_test)
x_test
```

```python
# importing cost function libraries
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
# fitting linear regression model
lm = LinearRegression()
lm.fit(x_train,y_train)
# regression coefficients
print('Coefficients: ', lm.coef_)
```

```python
# fitting model into test data
y_pred = lm.predict(x_train)
```

```python
# calculting rsquared value  # r2 score for train data
r2 = r2_score(y_true=y_train,y_pred=y_pred)
print(r2)
```

```python
y_test_pred=lm.predict(x_test) # predicting test set
```

```python
r2 = r2_score(y_test_pred,y_test) # r2 score for test data
print(r2)
```

```python
# Calculating error
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_test_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_test_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```
